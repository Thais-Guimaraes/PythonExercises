## Objetivo

Demonstrar a <b>importação de múltiplos arquivos</b> em csv e realizar <b>tratamento de valores ausentes.</b>

Os dados foram retirados do site: https://bdmep.inmet.gov.br/ referente ao ano de 2022

São <b>567</b> arquivos que serão <b>unificados em 1 único DataFrame do Pandas</b>

O objetivo será carregar para o pandas as seguintes colunas:
Data;
PRECIPITAÇÃO TOTAL, HORÁRIO (mm);
Regiao;	
UF;
Cidade


Os arquivos possuem um padrão nas nomenclaturas, exibindo as informações da Região, UF e Cidade exemplo:




```python
from IPython.display import Image
Image('imagens/arquivo.jpg')
```

O que será realizado abaixo: <br>
o loop "for" irá ler cada arquivo, na pasta especificada
e para cada arquivo, vamos extrair o nome do arquivo,realizando uma separação (split) a partir do underscore "_"
será realizada abertura e leitura de cada arquivo, passando o delimitador, o encoding e as colunas que queremos,
que nesse caso serão apenas "data" e "precipitação total", será necessário pular 8 linhas para pegar o header

depois criamos 3 colunas com valores ausentes para representar as colunas Regiao, UF e Cidade

e então atribuímos a estas colunas as posições correspondentes com índices 1,2 e 4 pegando do nome do arquivo

```python
import pandas as pd
import os
import glob
import re
import csv
import numpy as np
from datetime import datetime
#!pip install missingno
import missingno as msno 
from matplotlib import pyplot as plt
```

```python
lista_arquivos = []
for file in glob.glob("dados\*.csv"):
    nome_arquivo = os.path.basename(file).split('_') 
    with open(file) as fp:
        arquivo = pd.read_csv(fp, delimiter = ";", encoding= "latin-1",usecols =['Data','PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'], skiprows = 8)
        arquivo[['Regiao','UF', 'Cidade']] = np.nan
        arquivo[['Regiao','UF', 'Cidade']] = [nome_arquivo[1], nome_arquivo[2], nome_arquivo[4]]
        lista_arquivos.append(arquivo)
df = pd.concat(lista_arquivos,ignore_index= True)        
```

```python
df['Data'] = pd.to_datetime(df['Data'], dayfirst = True)
```

```python
df.rename(columns={"PRECIPITAÇÃO TOTAL, HORÁRIO (mm)": "PrecipitacaoTotal"}, inplace = True)
```

```python
df['PrecipitacaoTotal'].replace(",",".")
```

```python
df['PrecipitacaoTotal'] = df['PrecipitacaoTotal'].str.replace(",",".").astype(np.float64)
```
